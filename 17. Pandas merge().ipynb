```python
import pandas as pd
```

```python
df = pd.DataFrame({'ID': [1,2,3,4] , 'Class' :[9,10,11,12]})
```

```python
df
```

```python
df1 = pd.DataFrame({'ID': [1,2,3,4] , 'Name' :['Ankit' , 'Nitish' , 'Shubham' ,'Laxman']})
df1
```

```python
pd.merge(df,df1) #left right paramter
```

```python
pd.merge(df,df1 , on='ID')
```

```python
pd.merge(df1,df , on='ID')
```

```python
 df1 = pd.DataFrame({'ID': [1,2,3,5] , 'Name' :['Ankit' , 'Nitish' , 'Shubham' ,'Laxman']})
df1
```

```python
pd.merge(df1,df , on='ID') #if id is not same 
```

```python
#how paramter
```

```python
pd.merge(df1,df , on='ID' , how= 'inner')
```

```python
pd.merge(df,df1 , on='ID' , how= 'left')
```

```python
pd.merge(df1,df , on='ID' , how= 'left')
```

```python
pd.merge(df,df1 , on='ID' , how= 'right')
```

```python
pd.merge(df,df1 , on='ID' , how= 'outer')
```

```python
#indicator Paramter
pd.merge(df,df1 , on='ID' , how= 'outer' , indicator=True)
```

```python
 df1 = pd.DataFrame({'ID': [5,6,7,8] , 'Name' :['Ankit' , 'Nitish' , 'Shubham' ,'Laxman']})
df1
```

```python
 df2 = pd.DataFrame({'ID': [5,6,7,8] , 'Name' :['Ankit' , 'Nitish' , 'Shubham' ,'Laxman']})
df2
```

```python
pd.merge(df1,df2 , on='ID')
```

```python
#suffixes paramter 
```

```python
pd.merge(df1,df2 , on='ID' , suffixes= ('_Higher', '_Middle'))
```
