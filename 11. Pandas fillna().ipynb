# Value Paramter 

```python
df.fillna(0)  #first paramter Value set as '0 ' so all NAN will be filled by 0
```

```python
df.fillna({'Name' : 'none' })
```

# method Paramter

```python
df.fillna(method = 'ffill') #forward fill
```

```python
df.fillna(method = 'pad') #forward fill
```

```python
df.fillna(method = 'bfill') #backword fill
```

# axis paramter

```python
df.fillna(axis = 0) # row change  
```

```python
# with axis we have to use any one paramater i.e value or method
```

```python
df
```

```python
df.fillna(method = 'ffill', axis = 0) # row change  
```

```python
df.fillna(method = 'ffill', axis = 1) # column change  
```

```python
df.fillna(method = 'bfill', axis = 1) # column change  
```

# Limit Paramter 

```python
df.fillna(limit = 1 , value =0) # it will fill value at one place in colum as we have given limit 1
```

```python
df.fillna(limit = 2 , value =0)
```

```python
df.fillna(limit = 1 , method ='ffill')
```

```python
df.fillna(limit = 2 , method ='ffill')
```

# inplace

```python
df.fillna(inplace= True , value=5) # modify existing dataframe
df
```
